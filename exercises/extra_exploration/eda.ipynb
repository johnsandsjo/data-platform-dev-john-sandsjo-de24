# EDA on Crypto API

## Quotes Latest v2
endpoint = v2/cryptocurrency/quotes/latest

```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv('API_KEY')

url = "https://pro-api.coinmarketcap.com/v2/cryptocurrency/quotes/latest"

paramteres = {
    'symbol' : 'BTC',
    'convert' : 'SEK'
}

headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': API_KEY
}

response = requests.get(url, params=paramteres, headers=headers)
coin_data = response.json()["data"]
coin_data
```

```python
btc_data = coin_data['BTC'][0]
btc_data
```

```python
btc_data['name']
```

```python
quote = btc_data['quote']
quote
```

```python
price = quote['SEK']['price']
f"{price:.2f}"
```

## Listings latest V1
endpoint = v1/cryptocurrency/listings/latest

```python
listing_url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest"

listing_parameters = {
    'limit' : 10,
    'convert' : 'USD'
}

response = requests.get(listing_url, params=listing_parameters, headers=headers)
listing_data = response.json()["data"]
listing_data
```

```python
for currency in listing_data:
    print(currency["name"])
    quotes = currency["quote"]
    for key, value in quotes.items():
        print(f"Price = {value['price']:.2f} Volume = {value['volume_24h']:.2f}")
```
