## 1. Working with grades data

```python
import pandas as pd

df_dict = {}
for sheet in ["Tabell 1A", "Tabell 1B"]:
    df_dict[sheet] = pd.read_excel("data/betyg_o_prov_riksnivå.xlsx", sheet_name=sheet, skiprows=7)
df_dict["Tabell 1B"]
```

```python
#a) Draw a line chart for proportion of students that are missing approved grade for one or more subjects for the years 18-23.
import seaborn as sns
import matplotlib.pyplot as plt
years =  df_dict["Tabell 1B"]["Unnamed: 0"].head(5)
percent_failed= df_dict["Tabell 1B"]["Totalt.2"].head(5)

new_dict = {years[i] : percent_failed[i] for i in range(len(percent_failed))}
df = pd.DataFrame(index = new_dict.keys(), data= new_dict.values(), columns= ['percent_failed'])
sns.lineplot(df)
```

```python
new_df_dropped = df_dict["Tabell 1B"].drop(columns= ["Totalt", "Flickor", "Totalt.1", "Pojkar", "Flickor.1", "Pojkar.1", "Flickor.2", "Pojkar.2"])
new_df_dropped
```

```python
new_df_dropped_renamed = new_df_dropped.rename({
    "Unnamed: 0": "Kategori", 
    "Totalt.2": "percent_failed"},
    axis=1)
new_df_dropped_renamed
```

```python
table_to_query = new_df_dropped_renamed
query = """
SELECT 
    Kategori,
    percent_failed
FROM
    table_to_query
WHERE Kategori IN ('Förgymnasial utbildning', 'Gymnasial utbildning', 'Eftergymnasial utbildning')
"""
print(query)
```

```python
import duckdb

duckdb.query(query)
```

```python
import pandas as pd
df_os = pd.read_csv("data/athlete_events.csv")
df_os
```

```python
import duckdb
duckdb.query("SELECT * FROM df_os").df()
```

```python
#d) Find out the 10 oldest atheletes, their age and the sport
duckdb.query("""
             SELECT Name, Age, Sport FROM df_os ORDER BY AGE DESC LIMIT 10
             """).df()
```

```python
#d) Find out the 10 oldest atheletes, their age and the sport
duckdb.query("""
             SELECT Name, Age, Sport, Year FROM df_os ORDER BY AGE LIMIT 10
             """).df()
```

```python
#e) Find out the five sports with highest median age
duckdb.query("""
             SELECT Sport, MEDIAN(Age) as median_age FROM df_os GROUP BY Sport ORDER BY median_age DESC LIMIT 5
             """).df()
```

```python
#f) Find out the five sports with lowest median age
duckdb.query("""
             SELECT Sport, MEDIAN(Age) as median_age FROM df_os GROUP BY Sport ORDER BY median_age LIMIT 5
             """).df()
```

```python
#f) Find out top 10 countries after number of gold medals
duckdb.query("""
             SELECT NOC, count(Medal) as Num_of_golds FROM df_os WHERE Medal = 'Gold' group by NOC ORDER BY Num_of_golds DESC LIMIT 10
             """).df()
```

```python
#g) Find out top 10 countries after number of medals
duckdb.query("""
             SELECT NOC, count(Medal) as Num_of_golds FROM df_os group by NOC ORDER BY Num_of_golds DESC LIMIT 10
             """).df()
```
