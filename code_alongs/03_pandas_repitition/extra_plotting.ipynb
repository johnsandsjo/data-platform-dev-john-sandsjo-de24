# Plotting, basics
- Matplotlib
- Seaborn
- Pandas plot() method

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv')
df.head()
```

```python
df.info()
```

***matlap syntax in matplotlib***

```python
import matplotlib.pyplot as plt
plt.hist(df["age"], bins = 15)
plt.xlabel("age")
plt.ylabel("count")
plt.title("Histogram of ages in titanic dataset")
```

***OOP approach to matplotlib***

```python
fig, ax = plt.subplots(1)
ax.hist(df["age"], bins=15)
ax.set(title="Histogram of ages in titanic dataset", xlabel="age", ylabel="count")
```

***bar chart in matplotlib***

```python
df.columns
```

```python
classes = df["class"].value_counts().reset_index()
classes
```

```python
fig, ax = plt.subplots(1)
ax.bar(x=classes["class"], height = classes["count"], color = ["blue", "red", "yellow"])
ax.set(xlabel="class", ylabel="count", title="Classes on titanic")
```

```python
cohorts = ["DE23", "DE24", "AI23", "JAVA24"]
number_students = [29, 38, 23, 24]

fig, ax = plt.subplots(1)
ax.bar(x=cohorts, height=number_students)
```

****bar chart in seaborn****

```python
import seaborn as sns
classes
```

```python
ax = sns.barplot(classes, x="class", y="count")
ax.set(title= 'Classes onboard of titanic')

#skillnad seaborn och matplotlib. I matplotlib måste man speca mer. Seaborn kan bara ange dataframe och sedan vilka columns
```

```python
sns.histplot(df, x="age", hue="sex", multiple="dodge", kde= True)
```
